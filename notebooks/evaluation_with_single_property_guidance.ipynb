```python
REPO_PATH = ''

import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import MolFromSmiles, Draw
import random
import matplotlib
from lets_plot import *
LetsPlot.setup_html()
```

```python
# set the property you want to investigate (must match a column name of your data)
PROPERTY = 'mol_weight'
```

```python
# load the generated data
data = pd.read_csv(REPO_PATH + '/outputs/mol_weight_down_generation.csv')
```

```python
# create a set of colours for plotting
cmap = matplotlib.cm.get_cmap('Greens', len(set(data['guidance']))+4)   
colors = []
for i in range(cmap.N):
    rgba = cmap(i)
    # rgb2hex accepts rgb or rgba
    colors.append(matplotlib.colors.rgb2hex(rgba))
colors = colors[2:-2]
```

```python
LetsPlot.setup_html()
g = ggplot(data, aes(x=PROPERTY)) + \
        geom_density(aes(group='guidance', color='guidance', fill='guidance'),
                       alpha=0.2) + \
        ggtitle('Decreasing Molecular Weight') + \
        scale_color_manual(values = colors) + \
        scale_fill_manual(values = colors) + \
        xlab('Molecular Weight')
g
```
